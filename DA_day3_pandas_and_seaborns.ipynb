In build data sets in seaborn

*  To  load we use:load_dataset("dataset name")


*  hue=different color for different category
*  style=differnt shapes

*   color palette-pastal,bright,dark,muted,colorblind,deep

*   kind=line

```python
tips=sns.load_dataset("tips")
fmri=sns.load_dataset("fmri")
dowjones=sns.load_dataset("dowjones")
dots=sns.load_dataset("dots")
healthexp=sns.load_dataset("healthexp")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="day",style="time")
#hue - creating difference based on a column via colors
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="size",style="smoker",palette="dark")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="size",style="smoker",palette="colorblind")
```

```python
sns.relplot(data=dowjones,x="Date",y="Price",kind="line")
```

```python
sns.relplot(data=fmri,x="timepoint",y="signal")
```

```python
sns.catplot(data=tips,x="total_bill",y="day")
```

```python
sns.catplot(data=fmri,x="region",y="timepoint")
```
